 **Обработка routes_data**

```python
import pandas as pd

routes = pd.read_csv('routes_raw_data.csv')
```

```python
routes.head()
```

```python
tram_routes = routes[routes['Тип транспорта'] == 'Трамвай']
tram_routes = tram_routes[tram_routes['Направление движения'] == 1]
```

```python
tram_routes.head()
```

Составим нужную нам табличку

```python
useful_df = tram_routes.groupby('Идентификатор маршрута').agg({
    'Числовой номер маршрута': 'first',
    'Название остановки': 'nunique',
    'Расстояние до следующей остановки, км': 'sum'
}).reset_index()

useful_df.columns = ['Идентификатор маршрута', 'Маршрут', 'Количество остановок', 'Длина маршрута']
```

```python
sorted_df = useful_df.sort_values('Маршрут')
```

```python
sorted_df.to_csv('routes_stops_count_and_len.csv', index=False, encoding='utf-8-sig')
sorted_df.reset_index(drop=True, inplace=True)
```
