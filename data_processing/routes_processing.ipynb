 **Обработка routes_data**

```python
import pandas as pd

routes = pd.read_csv('routes_data.csv')
```

```python
routes.head()
```

```python
tram_routes = routes[routes['Тип транспорта'] == 'Трамвай']
tram_routes = tram_routes[tram_routes['Направление движения'] == 1]
```

```python
tram_routes.head()
```

Составим нужную нам табличку

```python
useful_df = tram_routes.groupby('Числовой номер маршрута').agg({
    'Название остановки': 'nunique',
    'Расстояние до следующей остановки, км': 'sum'
}).reset_index()

useful_df.columns = ['Маршрут', 'Количество остановок', 'Длина маршрута']
```

```python
useful_df
```

```python
useful_df.to_csv('routes.csv', encoding='utf-8-sig')
```
